```
#Hello World
#Create RDD 
txt = sc.textFile("/FileStore/tables/helloWorld.txt")

#map reduce
counts = txt.flatMap(lambda x:x.split(' ')).map(lambda x:(x,1)).reduceByKey(lambda x,y:x+y)

#action
output = counts.take(10)
for word, count in output:
  print("%s : %i" %(word ,count))
```

```
#Python Funtional Programming
#Map
def sqrFunc(ele):
  return (ele*ele)
myList = [1,2,3,4,5,6,7]
sqrList = map(sqrFunc, myList)

sqrListLambda = map(lambda x:x*x,myList)

print(list(sqrList))
print(list(sqrListLambda))

#Filter
evenList = filter(lambda x:x%2==0, myList)
print(list(evenList))

#Reduce
import functools
sum = functools.reduce(lambda x,y:x+y, myList)
print(sum)
```

```
#RDD
myRDD = sc.parallelize(myList)

#Map
sqrRDD = myRDD.map(lambda x:x*x)
sqrRDD.collect()

#Filter
evenRDD = myRDD.filter(lambda x:x%2==0)
evenRDD.collect()
```
